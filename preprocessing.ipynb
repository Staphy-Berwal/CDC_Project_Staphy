```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

target = 'log_price'
drop_cols = ['id','date','price','log_price']

X_tab = df.drop(columns=drop_cols)
y = df[target]

X_train, X_val, y_train, y_val = train_test_split(
    X_tab, y, test_size=0.2, random_state=42
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score

model_tab = Ridge(alpha=1.0)
model_tab.fit(X_train_scaled, y_train)

y_pred = model_tab.predict(X_val_scaled)

rmse = np.sqrt(mean_squared_error(y_val, y_pred))
r2 = r2_score(y_val, y_pred)

print("Tabular RMSE:", rmse)
print("Tabular R:", r2)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

xgb_model = XGBRegressor(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='reg:squarederror',
    random_state=42
)

xgb_model.fit(X_train, y_train)

y_pred_xgb = xgb_model.predict(X_val)

rmse_xgb = np.sqrt(mean_squared_error(y_val, y_pred_xgb))
r2_xgb = r2_score(y_val, y_pred_xgb)

print("XGBoost RMSE:", rmse_xgb)
print("XGBoost R:", r2_xgb)
```

```python
importance = xgb_model.feature_importances_
features = X_train.columns

imp_df = pd.DataFrame({
    'feature': features,
    'importance': importance
}).sort_values(by='importance', ascending=False)

imp_df.head(10)
```
